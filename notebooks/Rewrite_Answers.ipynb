```python
import json
import pandas as pd
import numpy as np
```

```python
#read in the raw data
seqdataIn = pd.read_json('../data/seq-private.json', orient='records')
print(seqdataIn)
```

```python
sequences = seqdataIn['stem']
ids = seqdataIn['id']
s = dict(zip(ids,sequences))
print(s)
```

```python
#overwrite
with open('../data/seq-private.answer.json') as f:
    j = json.load(f)
print(j)
```

```python
nj = {}
for i in range(len(ids)):
    if str(i) in j.keys():
        nj[str(ids[i])] = j[str(i)]
nj
```

```python
with open('../data/seq-private2.answer.json', 'w', encoding='utf-8') as f:
    json.dump(nj, f)
```
